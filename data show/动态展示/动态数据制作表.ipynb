# 数据导出

```python
import pymongo
from pymongo.collection import Collection
from datetime import datetime
#建立连接
client=pymongo.MongoClient('localhost',27017)
#建立数据库
db=client["bilibili"]

#从原有的txt文件导入share_id：
def find(end_time):
    #表的对象化
    mgtable=Collection(db,'mbilibili_22')

    info=mgtable.find({"create":{"$gte":datetime(2012,1,1),"$lte":datetime(end_time[0],end_time[1],end_time[2])}})
    
    like_list=[]
    for i in info:
        like_list.append((i['author'],i['play'],i['create']))

    return like_list

end_time=[]
for year in range(2012,2020):
    for m in range(1,13):
        if year==2019 and m ==6:
            break
        end_time.append((year,m,1))

up_data=[]
for end in end_time:
    like_list=find(end)
    if like_list !=[]:
        up={}
        
        for i in like_list:
            if i[0] in up:
                try:
                    if i[1] != '--':
                        up[i[0]][0]+=i[1]
                        
                except:
                    pass
            else:
                if i[1] != '--':
                    up[i[0]]=[int(i[1]),'{}-{}-{}'.format(end[0],end[1],end[2])]
                    
# up
        
        
        q=sorted(up.items(),key=lambda item:item[1][0],reverse=True)

        up_data.append(q[:20])
up_data
        


```

# 生成xls文件

```python
# -*- coding: utf-8 -*-
import xlrd
import xlwt
from datetime import date, datetime


up=[]
value=[]
date=[]
for data in up_data:
    for sett in data:
        up.append(sett[0])
        value.append(sett[1][0])
        date.append(sett[1][1])

def write_excel():
    f = xlwt.Workbook() #创建工作簿

    sheet1 = f.add_sheet(u'sheet1') #创建sheet
    row0 = [u'name',u'value',u'data']
    column0 = up
    column1 = value
    column2 = date


    #生成第一行
    for i in range(0,len(row0)):
        sheet1.write(0,i,row0[i])

    #生成第二列
    i = 0
    for j in range(0,len(column0)):
        sheet1.write(i+1,0,column0[j])
        i += 1
    #生成第二列
    i = 0
    for j in range(0,len(column1)):
        sheet1.write(i+1,1,column1[j])
        i += 1

    #生成第二列
    i = 0
    for j in range(0,len(column2)):
        sheet1.write(i+1,2,column2[j])
        i += 1

    f.save('mbilibili_176.xls') #保存文件
if __name__ == '__main__':

    write_excel()
```
