## 点赞前一百up主

```python
import pymongo
import matplotlib.pyplot as plt
from pymongo.collection import Collection
import matplotlib
plt.style.use('ggplot')
matplotlib.rcParams['font.family']='SimHei'
matplotlib.rcParams['font.style']='normal'
matplotlib.rcParams['font.size']=7

#建立连接
client=pymongo.MongoClient('localhost',27017)
#建立数据库
db=client["bilibili"]

#从原有的txt文件导入share_id：
def find(code):
    #表的对象化
    mgtable=Collection(db,'mbilibili_{}'.format(code))

    info=mgtable.find({"like":{ "$gt":0} } ).sort([("like",-1)]).limit(100)
    
    like_list=[]
    for i in info:
        
        like_list.append(i['author'])

    return like_list
like_list=find('游戏-单机游戏')
author_dict={}
# author=()
for author in like_list:
    if author in author_dict:
        author_dict[author]+=1
    else:
        author_dict[author]=1

q=sorted(author_dict .items(),key=lambda item:item[1])

b=plt.barh([x[0] for x in q], [y[1] for y in q])

for rect in b:
    w=rect.get_width()
    plt.text(w,rect.get_y()+rect.get_height()/2,'%d'%int(w),ha='left',va='center')

# plt.xlim((0,13000000))
# plt.xticks([x*1000000 for x in range(0,14)])
plt.ylabel('up主')
plt.xlabel('视频数量')
plt.xticks(fontsize=8)
plt.yticks(fontsize=6)
plt.title('单机游戏区点赞前一百视频up主')
# plt.show()
plt.savefig('big.png',dpi=300,bbox_inches='tight')
```

## 播放前20视频

```python
import pymongo
import matplotlib.pyplot as plt
from pymongo.collection import Collection
import matplotlib
plt.style.use('ggplot')
matplotlib.rcParams['font.family']='SimHei'
matplotlib.rcParams['font.style']='normal'
matplotlib.rcParams['font.size']=7

#建立连接
client=pymongo.MongoClient('localhost',27017)
#建立数据库
db=client["bilibili"]

#从原有的txt文件导入share_id：
def find(code):
    #表的对象化
    mgtable=Collection(db,'mbilibili_{}'.format(code))

    info=mgtable.find({"play":{ "$gt":0} } ).sort([("play",-1)]).limit(20)
    
    like_list=[]
    for i in info:
        
        like_list.append((i['title'],i['play']))

    return like_list
like_list=find('游戏-单机游戏')
```

```python
like_list.reverse()

b=plt.barh([x[0] for x in like_list], [y[1] for y in like_list])

for rect in b:
    w=rect.get_width()
    plt.text(w,rect.get_y()+rect.get_height()/2,'%d'%int(w),ha='left',va='center')

# plt.xlim((0,13000000))
# plt.xticks([x*1000000 for x in range(0,14)])
plt.ylabel('up主')
plt.xlabel('视频数量')
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.title('单机游戏区播放前20视频')
# plt.show()
plt.savefig('Top20.png',dpi=300,bbox_inches='tight')
```
